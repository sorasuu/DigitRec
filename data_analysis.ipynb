```python
# Import the nessecery library 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import the data
data = pd.read_csv('data/semeion.data', delimiter='\s+')
```

```python
# get the label from the data so that we can later train them
def label_transform(df):
    ls= df.values
    label=0
    for i in ls:
        if i==1:
            break
        elif i==0:
            label+=1
        else:
            print("wrong format")
    return label
```

```python
# Implement our preprocess to the data
label_list = []
data_list = []
data_label = data.iloc[:,256:266]
for i in range(len(data_label.index)):
    label = label_transform(data_label.loc[i])
    label_list.append(label)
data_list=data.drop(data_label.columns,axis=1).values
label_list = np.array(label_list)
```

----------------------------------------------------------------------------------------------------------------------------------------
***PCA decompostion***

```python
plt.figure()
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
proj = pca.fit_transform(data_list)
plt.scatter(proj[:, 0], proj[:, 1], c= label_list , cmap="Paired")
plt.title('PCA linear transformation')
plt.colorbar()
```

-----------------------------------------------------------------------------------------------------------------------------------------
***Distribution of the dataset***

```python
label_list = pd.DataFrame(label_list)
plt.figure()
label_list.hist()
plt.title('Data Distribution')
plt.show()
```
